```python
import h5py

def load_hdf5(db_file):
    with h5py.File(db_file, 'r') as f:
        train_x, train_y = f['train_x'][...], f['train_y'][...]
        test_x, test_y = f['test_x'][...], f['test_y'][...]
        
        return train_x, train_y, test_x, test_y
```

```python
train_x, train_y, test_x, test_y = load_hdf5('data.h5')
```

```python
import matplotlib.pyplot as plt

img = train_x[0]
plt.imshow(img)
plt.show()
```

```python
import numpy as np

'''
Transform shape of data from (250, 64, 64, 3) to (12288, 250)
- represent each image as column vector ! (12288, 1)
'''

flat_train_x = np.zeros(shape=(np.prod(train_x.shape[1:]), train_x.shape[0]))
flat_test_x = np.zeros(shape=(np.prod(test_x.shape[1:]), test_x.shape[0]))

for i, sample in enumerate(train_x):
    flat_train_x[:, i] = sample.flatten()
    
for i, sample in enumerate(test_x):
    flat_test_x[:, i] = sample.flatten()
    
flat_train_x /= 255.0
flat_test_x /= 255.0
```

```python
'''
Wersja z iterowaniem po przykładach uczących w pętli
(wyodrębnianie obrazka : sample)
'''

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def loss(output, y):
    return -np.log(1-output)*(1-y) + -np.log(output)*y

```

```python
'''
Regresja logistyczna z iterowaniem po kolejnych przykładach uczących w pętli.
'''

weights = np.zeros(shape=(flat_train_x.shape[0]), dtype=np.float32)
bias = 0.0
alpha = 0.01
num_iters = 1500
for x in range(num_iters):
    
    J = 0.0
    w_delta = np.zeros(shape=weights.shape, dtype=np.float32)
    b_delta = 0.0

    for i in range(flat_train_x.shape[1]):
        sample = flat_train_x[:, i]
        label = train_y[i]
        output = sigmoid(np.dot(weights, sample) + bias)
        J += loss(output, label)
        delta = -alpha * (output - label)

        w_delta += sample * delta
        b_delta += delta        
        
    
    if x % 500 == 0:
        print(x+1, ":", J, J/flat_train_x.shape[1]); print()
    
    weights += w_delta / flat_train_x.shape[1]
    bias += b_delta / flat_train_x.shape[1]
    
    
```

```python
from sklearn.metrics import confusion_matrix

'''
Regresja logistyczna w postaci wektorowej
(brak pętli po przykładach uczących, operacje na macierzach)
'''

num_iters = 1500
learning_rates = [0.01, 0.005, 0.0025]
results = dict([(alpha, []) for alpha in learning_rates])

for alpha in learning_rates:
    
    weights = np.zeros(shape=(flat_train_x.shape[0]), dtype=np.float32)
    bias = 0.0
    print("\n>>> Learning rate {}:".format(str(alpha)))
    
    for i in range(num_iters):
        outputs = sigmoid(np.dot(weights, flat_train_x) + bias)
        
        delta = -alpha * (outputs - train_y)
        w_delta = np.sum(flat_train_x * delta, axis=1) 
        
        J = loss(outputs, train_y).mean()
            
        if i % 100 == 0:
            results[alpha].append(J)

        weights += w_delta / flat_train_x.shape[1]
        bias += np.sum(delta) / flat_train_x.shape[1]
                
    # check training set    
    outputs = sigmoid(np.dot(weights, flat_train_x) + bias)
    J = loss(outputs, train_y).mean()
    
    outputs[outputs < 0.5] = 0.0
    outputs[outputs >= 0.5] = 1.0
    
    cf = confusion_matrix(train_y, outputs)
    tn, fp, fn, tp = cf.ravel()
    print('\nTRAIN SET\nConfusion matrix:\n{}\nAccuracy: {}, Loss: {}'.format(cf, (tp+tn)/(np.sum(cf)), J))
    
    # check test set
    outputs = sigmoid(np.dot(weights, flat_test_x) + bias)
    J = loss(outputs, test_y).mean()
    
    outputs[outputs < 0.5] = 0.0
    outputs[outputs >= 0.5] = 1.0
    
    cf = confusion_matrix(test_y, outputs)
    tn, fp, fn, tp = cf.ravel()
    print('\nTEST SET\nConfusion matrix:\n{}\nAccuracy: {}, Loss: {}'.format(cf, (tp+tn)/(np.sum(cf)), J))
    
```

```python
'''
Analiza zbieżności algorytmu dla różnych prędkości uczenia.
'''

import matplotlib.pyplot as plt

for alpha in learning_rates:
    plt.plot(np.arange(len(results[alpha])), results[alpha])

plt.xlabel('Iterations (x100)')
plt.ylabel('Cost')
plt.title(r'Learning rate $\alpha$')
plt.legend(learning_rates)
plt.show()
```

```python
def initialize(dim):
    weights = np.zeros(shape=dim, dtype=np.float32)
    bias = 0.0
    return weights, bias
    
def propagate(w, b, X, Y):
    
    outputs = sigmoid(np.dot(w, X) + b)
    db = np.sum(outputs - Y)
    dw = np.sum(X * db, axis=1) 

    cost = loss(outputs, train_y).mean()
    grads = {'dw' : dw, 'db' : np.sum(db)}
    
    return grads, cost

def optimize(w, b, X, Y, num_iters, alpha):
    costs = []
    
    for i in range(num_iters):
        grads, cost = propagate(w, b, X, Y)
        dw = grads['dw']
        db = grads['db']
        w += dw / X.shape[1]
        b += db / X.shape[1]
        if i % 100 == 0:
            costs.append(cost)
        
    params = {'w' : w, 'b' : b}
    return params, costs

def model(X_train, Y_train, X_test, Y_test, num_iters, alpha):
    w, b = initialize(X_train.shape[0])
    
    # Gradient descent
    params, costs = optimize(w, b, X_train, Y_train, num_iters, alpha)
    
    # Retrieve parameters w and b
    w = params["w"]
    b = params["b"]
    
    # Predict test/train examples
    train_pred = predict(w, b, X_train)
    test_pred = predict(w, b, X_test)
    
    # Print train/test accuracy
    d = {"costs": costs,
        "w" : w,
        "b" : b}
    
    return d
```

```python
'''
Zmienna models to lista modeli wraz z ich parametrami oraz dodatkowo 
listą z wartościami funkcji kosztu wraz z biegiem iteracji.
'''

models = []
```

```python
# TANGENS HIPERBOLICZNY! w warstwie ukrytej
# lub: ReLU! jeszcze lepiej
# na wyjściu: sigmoida

'''
Sieć neuronowa z jedną warstwą ukrytą (hidden size - liczba neuronów w tej warstwie)
'''

np.random.seed(1)

input_size = flat_train_x.shape[0]
hidden_size = 50
output_size = 1

m = flat_train_x.shape[1]

learning_rates = [0.005, 0.01, 0.05, 0.1]
results = dict([(alpha, []) for alpha in learning_rates])

for alpha in learning_rates:
    print(); print(alpha)
    
    W1 = np.random.randn(hidden_size, input_size) * 0.1
    W2 = np.random.randn(output_size, hidden_size) * 0.1

    b1 = np.zeros(shape=(hidden_size, 1), dtype=np.float32)
    b2 = np.zeros(shape=(output_size, 1), dtype=np.float32)
    
    for i in range(10000):

        Z1 = W1.dot(flat_train_x) + b1
        A1 = sigmoid(Z1)

        Z2 = W2.dot(A1) + b2
        A2 = sigmoid(Z2)

        if i % 100 == 0:
            J = np.average(loss(A2, train_y))
            results[alpha].append(J)
            if i % 1000 == 0:
                print(J)
        
        dA2 = A2 - train_y
        dZ2 = dA2
        
        dW2 = np.dot(dZ2, A1.T) * (1/m)
        db2 = np.sum(dZ2) * (1/m)
                
        dA1 = np.dot(W2.T, dZ2)
        dZ1 = dA1 * (A1 * (1-A1))
        
        dW1 =  np.dot(dZ1, flat_train_x.T) * (1/m)
        db1 = np.sum(dZ1) * (1/m)
        
        W2 += -alpha * dW2
        b2 += -alpha * db2

        W1 += -alpha * dW1
        b1 += -alpha * db1
        
    models.append({
        'alpha' : alpha,
        'hidden_units' : hidden_size,
        'W1' : W1, 'b1' : b1, 'W2' : W2, 'b2' : b2,
        'result' : results[alpha]
    })
```

```python
'''
Wykres funkcji kosztu dla różnych prędkości uczenia.
'''

for alpha in learning_rates:
    plt.plot(np.arange(len(results[alpha])), results[alpha])

plt.xlabel('Iterations (x100)')
plt.ylabel('Cost')
plt.title(r'Hidden Layer Size  $n_h$ = {}'.format(hidden_size))
plt.legend(learning_rates)
# plt.show()

plt.savefig('{}units.pdf'.format(hidden_size))
```

```python
max_accuracy = 0.0
best_model = None

for model in models:
    print('>>> Learning rate: {}, hidden units: {}'.format(model['alpha'], model['hidden_units']))
    
    print("\nTRAIN SET")
    Z1 = model['W1'].dot(flat_train_x) + model['b1']
    A1 = sigmoid(Z1)

    Z2 = model['W2'].dot(A1) + model['b2']
    A2 = sigmoid(Z2)
    
    acc, J = analyze(A2.T, train_y)
    if acc < 0.75:
        continue
    
    print("\nTEST SET")
    Z1 = model['W1'].dot(flat_test_x) + model['b1']
    A1 = sigmoid(Z1)

    Z2 = model['W2'].dot(A1) + model['b2']
    A2 = sigmoid(Z2)
    
    acc, J = analyze(A2.T, test_y)
    
    if acc > max_accuracy:
        max_accuracy = acc
        best_model = model
    
    print()
```

```python
from sklearn.metrics import confusion_matrix

'''
Funkcja obliczająca skuteczność (accuracy) oraz koszt, a także wyświetlająca macierz pomyłek dla danej predykcji.
'''

def analyze(output, y):
    
    J = loss(output, y).mean()
    
    output[output < 0.5] = 0.0
    output[output >= 0.5] = 1.0
    
    cf = confusion_matrix(y, output)
    tn, fp, fn, tp = cf.ravel()
    acc = (tp+tn)/np.sum(cf)
    print('Confusion matrix:\n{}\nAccuracy: {}, Loss: {}'.format(cf, acc, J))
    
    return acc, J
    
```
